```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Model names
model_names = ['VGG16', 'VGG19','Resnet', 'MobileNetV2', 'InceptionV3', 'DenseNet169']


#Various Metrics
test_loss = [0.193, 0.279, 0.324, 0.49, 0.426, 0.304]
test_accuracy = [0.927, 0.897, 0.971, 0.581, 0.858, 0.896]
test_auc = [0.993, 0.986, 0.982, 0.842, 0.974, 0.981]
test_precision = [0.928, 0.898, 0.903, 0.620, 0.867, 0.899]
test_recall = [0.924, 0.891, 0.97, 0.483, 0.852, 0.889]
```

```python
# Set the style of the plots
plt.style.use('ggplot')
```

```python
plt.figure(figsize=(15, 6))

# Bar plot for accuracy
colors = ['skyblue', 'lightgreen', 'salmon', 'gold', 'plum', 'teal', 'tomato','cyan']
plt.bar(model_names, test_accuracy,color=colors)
plt.plot(model_names, test_accuracy,marker='o',color='red')

# Adding labels and title
plt.xlabel('Model Names',fontsize=12, fontweight='bold')
plt.ylabel('Accuracy',fontsize=12, fontweight='bold')
plt.title('Accuracy Plot',fontsize=12, fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.5)

label_offset = max(test_accuracy) * 0.005

for i, v in enumerate(test_accuracy):
    plt.text(i,  v + label_offset,str(round(v,2)), ha='center', va='bottom', fontsize=10, fontweight='bold')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(15, 6))

# Bar plot for loss
colors = ['skyblue', 'lightgreen', 'salmon', 'gold', 'plum', 'teal', 'tomato','cyan']
plt.bar(model_names, test_loss,color=colors)
plt.plot(model_names, test_loss,marker='o',color='red')

# Adding labels and title
plt.xlabel('Model Names',fontsize=12, fontweight='bold')
plt.ylabel('Loss',fontsize=12, fontweight='bold')
plt.title('Loss Plot',fontsize=12, fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.5)

label_offset = max(test_loss) * 0.005

for i, v in enumerate(test_loss):
    plt.text(i,  v + label_offset,str(round(v,2)), ha='center', va='bottom', fontsize=10, fontweight='bold')

# Display the plot
plt.show()

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Sample true and predicted labels
true_labels = [
    'Mild_Demented', 'Very_Mild_Demented', 'Non_Demented',
    'Mild_Demented', 'Moderate_Demented', 'Non_Demented',
    'Very_Mild_Demented', 'Mild_Demented', 'Moderate_Demented',
    'Mild_Demented', 'Non_Demented', 'Very_Mild_Demented',
    'Non_Demented', 'Moderate_Demented', 'Very_Mild_Demented'
]

predicted_labels = [
    'Mild_Demented', 'Mild_Demented', 'Non_Demented',
    'Mild_Demented', 'Non_Demented', 'Non_Demented',
    'Very_Mild_Demented', 'Very_Mild_Demented', 'Moderate_Demented',
    'Mild_Demented', 'Non_Demented', 'Very_Mild_Demented',
    'Non_Demented', 'Very_Mild_Demented', 'Moderate_Demented'
]

# Define the classes
classes = ['Mild_Demented', 'Moderate_Demented', 'Non_Demented', 'Very_Mild_Demented']

# Create confusion matrix
cm = confusion_matrix(true_labels, predicted_labels, labels=classes)

# Plotting the confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=classes, yticklabels=classes)

plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
